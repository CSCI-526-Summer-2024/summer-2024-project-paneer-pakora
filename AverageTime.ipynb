```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = '/Users/sathyarajagopal/Downloads/event_data.json'
data = pd.read_json(file_path)


data['collectInsertedTimestamp'] = pd.to_datetime(data['collectInsertedTimestamp'])
data['eventDate'] = pd.to_datetime(data['eventDate'])
data['eventTimestamp'] = pd.to_datetime(data['eventTimestamp'])
data['gaUserStartDate'] = pd.to_datetime(data['gaUserStartDate'])


avg_time_taken_level_1 = data[data['eventLevel'] == 0]['timeTaken'].mean()


plt.figure(figsize=(6, 4))
plt.bar(['Level 1'], [avg_time_taken_level_1], color='teal', width=0.3)  
plt.xlabel('Level')
plt.ylabel('Average Time Taken (seconds)')
plt.title('Average Time Taken to Clear Level')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
